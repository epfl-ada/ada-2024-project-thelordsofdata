```python
#Imports
import numpy as np
import pandas as pd
import ast
import re
import matplotlib.pyplot as plt
import seaborn as sns
import os
import math
from statsmodels.stats import diagnostic
from scipy import stats
```

```python
df = pd.read_csv('data/almost_final.csv')
```

```python
df.head()
```

```python
df_usa = df[df['Countries'].str.contains('united states', na=False)]
print(len(df_usa))
```

```python
df_weird = df_usa[df_usa['domestic_percentage']<0].index
print(len(df_weird))
df_weird.head()
```

```python
sns.lmplot(x='Budget',y='domestic_percentage', data=df_usa)

print("Pearson coeff on budget and domestic")
print(stats.pearsonr(df_usa['Budget'],df_usa['domestic_percentage']))

sns.lmplot(x='Budget',y='foreign_percentage', data=df_usa)
print("Pearson coeff on budget and foreign")
print(stats.pearsonr(df_usa['Budget'],df_usa['foreign_percentage']))
```
