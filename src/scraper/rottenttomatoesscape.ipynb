```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_csv("../data/miscellaneous/almost_final_cleaned.csv")
df.head()
```

```python
print(len(df))
```

```python
import re
name_link = "you've got mail"
name_link = re.sub(r'[^a-zA-Z0-9 ]', '', name_link)
print(name_link)
```

```python
init_link = 'https://www.rottentomatoes.com/m/'

for index, row in df.iterrows():
    name_link = row['Name'].lower()
    name_link = re.sub(r'[^a-zA-Z0-9 ]', '', name_link)
    name_link = name_link.replace(" ", "_")
    name_link_year = name_link + "_" + str(row['Year'])
    try:
        r = requests.get(init_link + name_link_year)
        print(init_link + name_link_year)
        print(r.status_code)
        if r.status_code==404:

            raise TypeError("Bad Link")
    except:
        try:
            r = requests.get(init_link + name_link)
            print(init_link + name_link)
            print(r.status_code)
            if r.status_code==404:
                continue
        except:
            print("Can't find this movie")
            continue
    
    soup = BeautifulSoup(r.text, 'html.parser')
    audience = soup.find('rt-text', slot='audienceScore').text
    critics = soup.find('rt-text', slot='criticsScore').text
    #print(audience)
    df.loc[index, 'Audience_Score'] = audience
    #print(critics)
    df.loc[index, 'Critics_Score'] = critics
    #break
df.head()
```

```python
df.to_csv('./data/miscellaneous/rottentomatoes2.csv', index=False)
```
